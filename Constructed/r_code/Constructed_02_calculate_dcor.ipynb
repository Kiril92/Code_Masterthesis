```r
# Read raw data
filename_rawdata <- "../data/combined_data/constructed_images_combined_60_2022_08_25.csv"

rawdata <- read.csv(file = filename_rawdata, header = F)
```

```r
# Cast data from dataframe to matrix for the following function call
X <- data.matrix(rawdata)

# Calculate the distance correlations
corr_matrix <- my_dcor(X, logfile = "/home/jovyan/my_log.txt")

# Nan's to 0
corr_matrix_c <- corr_matrix
corr_matrix_c[is.na(corr_matrix_c)] <- 0
```
