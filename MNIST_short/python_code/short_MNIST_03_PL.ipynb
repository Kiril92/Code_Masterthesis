### Einlesen der Persistenzen

```python
# Get filenames from persistence-files of every subject
all_persistence_files = []

folderpath_persistences = "../data/persistences_from_filtrations/"

for file in glob.glob(folderpath_persistences + "*.pkl"):
    all_persistence_files.append(file)

all_persistence_files.sort()

print("No. of files:", len(all_persistence_files))
```

```python
# Read in all persistences

all_persistences = []

for i in range(len(all_persistence_files)):
    if i%25==0:
        print("Read file no.", i)
    
    filepath = all_persistence_files[i]
    
    with open(filepath, 'rb') as f:
        tmp = dill.load(f)
    all_persistences.insert(i, tmp)
```

# Calculate and save all persistence landscapes

```python
### PL-Parameter

# Tiefe der PL
pl_num_landscapes = 10

# Aufl√∂sung der PL
pl_resolution = 250

folderpath_avgPL = "../data/persistence_landscapes_averages/"
```

### H0

```python
avgPL_bucket01_H0_scaledWithin = calc_avg_PL_from_all_persistences(all_persistences = all_persistences,
                                                                   homology = 0,
                                                                   resolution = pl_resolution,
                                                                   num_landscapes = pl_num_landscapes,
                                                                   scaling = "within_subjects")

save_file(file = folderpath_avgPL + "avgPL_bucket01_H0_scaledWithin_" + str(date.today()) + ".pkl",
          x = avgPL_bucket01_H0_scaledWithin)
```

```python
avgPL_bucket01_H0_scaledBetween = calc_avg_PL_from_all_persistences(all_persistences = all_persistences,
                                                                    homology = 0,
                                                                    resolution = pl_resolution,
                                                                    num_landscapes = pl_num_landscapes,
                                                                    scaling = "between_subjects")

save_file(file = folderpath_avgPL + "avgPL_bucket01_H0_scaledBetween_" + str(date.today()) + ".pkl", 
         x = avgPL_bucket01_H0_scaledBetween)
```

```python
avgPL_bucket01_H0_unscaled = calc_avg_PL_from_all_persistences(all_persistences = all_persistences,
                                                               homology = 0,
                                                               resolution = pl_resolution,
                                                               num_landscapes = pl_num_landscapes,
                                                               scaling = "unscaled")

save_file(file = folderpath_avgPL + "avgPL_bucket01_H0_unscaled_" + str(date.today()) + ".pkl", 
          x = avgPL_bucket01_H0_unscaled)
```
