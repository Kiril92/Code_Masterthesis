### PL-Parameter

```python
# Anzahl der Buckets je Sample
pl_n_buckets = 3

# Anzahl der Persistence Landscapes je Bucket
pl_n_per_bucket = 5

# Tiefe der PL
pl_num_landscapes = 10

# Auflösung der PL
pl_resolution = 1000

# Sollen die Buckets in sich zusammengefügt werden?
concat_buckets = True
```

### Einlesen der Persistenzen

```python
# Get filenames from persistence-files of every subject
all_persistence_files = []

for file in glob.glob("/home/jovyan/rna/data/persistences/*.pkl"):
    all_persistence_files.append(file)

all_persistence_files.sort()

print("No. of files:", len(all_persistence_files))
```

```python
all_persistences = []

for i in range(len(all_persistence_files)):
    if i%25==0:
        print("Read file no.", i)
    
    filepath = all_persistence_files[i]
    
    with open(filepath, 'rb') as f:
        tmp = dill.load(f)
    all_persistences.insert(i, tmp)
```

### Aufteilung der  Persistenzen in Buckets

```python
all_persistences_in_buckets05 = seperate_into_buckets(all_persistences, 
                                                      bucket_size = 5,
                                                      verbose = True)
```

```python
all_persistences_in_buckets20 = seperate_into_buckets(all_persistences,
                                                      bucket_size = 20,
                                                      verbose = True)
```

### Persistenzdiagramme und PL (einzeln)

```python
subject = 400
```

```python
fig, ax = plt.subplots(5,2, figsize=(11, 18))

# Persistence diagrams
gd.plot_persistence_diagram(extract_homology(all_persistences[subject][0]), axes=ax[0,0])
gd.plot_persistence_diagram(extract_homology(all_persistences[subject][1]), axes=ax[1,0])
gd.plot_persistence_diagram(extract_homology(all_persistences[subject][2]), axes=ax[2,0])
gd.plot_persistence_diagram(extract_homology(all_persistences[subject][3]), axes=ax[3,0])
gd.plot_persistence_diagram(extract_homology(all_persistences[subject][4]), axes=ax[4,0])

# Persistence Landscapes
plot_pl(pl = pl_from_persistence(all_persistences[subject][0], 
                                 resolution = pl_resolution, 
                                 num_landscapes = pl_num_landscapes, 
                                 homology = 1),
        axes = ax[0, 1],
        resolution = pl_resolution,
        num_landscapes = pl_num_landscapes)

plot_pl(pl = pl_from_persistence(all_persistences[subject][1], 
                                 resolution = pl_resolution, 
                                 num_landscapes = pl_num_landscapes, 
                                 homology = 1),
        axes = ax[1, 1],
        resolution = pl_resolution,
        num_landscapes = pl_num_landscapes)

plot_pl(pl = pl_from_persistence(all_persistences[subject][2], 
                                 resolution = pl_resolution, 
                                 num_landscapes = pl_num_landscapes, 
                                 homology = 1),
        axes = ax[2, 1],
        resolution = pl_resolution,
        num_landscapes = pl_num_landscapes)

plot_pl(pl = pl_from_persistence(all_persistences[subject][3], 
                                 resolution = pl_resolution, 
                                 num_landscapes = pl_num_landscapes, 
                                 homology = 1),
        axes = ax[3, 1],
        resolution = pl_resolution,
        num_landscapes = pl_num_landscapes)

plot_pl(pl = pl_from_persistence(all_persistences[subject][4], 
                                 resolution = pl_resolution, 
                                 num_landscapes = pl_num_landscapes, 
                                 homology = 1),
        axes = ax[4, 1],
        resolution = pl_resolution,
        num_landscapes = pl_num_landscapes)

plt.tight_layout()  # Optional 
```

### Gemittelter PL

```python
# Gemittelter Landscape aus einem Bucket
L_avg = np.mean([pl_from_persistence(all_persistences[subject][0], resolution = pl_resolution, num_landscapes = pl_num_landscapes),
                 pl_from_persistence(all_persistences[subject][1], resolution = pl_resolution, num_landscapes = pl_num_landscapes),
                 pl_from_persistence(all_persistences[subject][2], resolution = pl_resolution, num_landscapes = pl_num_landscapes),
                 pl_from_persistence(all_persistences[subject][3], resolution = pl_resolution, num_landscapes = pl_num_landscapes),
                 pl_from_persistence(all_persistences[subject][4], resolution = pl_resolution, num_landscapes = pl_num_landscapes)],
                axis=0)

plot_pl(pl = L_avg,
        resolution = pl_resolution,
        num_landscapes = pl_num_landscapes)
```
